```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the dataset
data = pd.read_csv('GTT.csv')

# Extracting the relevant columns
X = data[['YEAR']]
y = data['TAVG']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate years for plotting (actual and predicted)
years_all = np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3)

# Predictions for all years
y_pred_all = model.predict(years_all.reshape(-1, 1))

# Plotting time series for actual and predicted data
plt.figure(figsize=(12, 8))

# Actual data
plt.plot(data['YEAR'], data['TAVG'], label='Actual', color='blue')

# Predicted data
plt.plot(years_all, y_pred_all, label='Predicted', color='red')

# Customize x-axis tick marks
plt.xticks(np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3))

plt.title('Average Temp. Vs Year - Prediction Set')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the dataset
data = pd.read_csv('GTT.csv')

# Extracting the relevant columns
X = data[['YEAR']]
y = data['TAVG']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Generate years for plotting (actual and forecasted)
years_all = np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3)
forecast_years = np.arange(data['YEAR'].max() + 1, 2031, 3)  # Forecast up to 2030 only with 3-year intervals

# Predictions for all years
y_pred_all = model.predict(years_all.reshape(-1, 1))
forecast_temps = model.predict(forecast_years.reshape(-1, 1))

# Plotting time series for actual and forecasted data
plt.figure(figsize=(12, 8))

# Actual data
actual_years = np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3)
plt.plot(data['YEAR'], data['TAVG'], label='Actual', color='blue')

# Forecasted data
forecasted_years = np.arange(data['YEAR'].max() + 1, 2031, 3)  # Adjusted to match the actual interval
plt.plot(forecasted_years, forecast_temps, label='Forecasted', color='green')

# Customize x-axis tick marks and grids for both actual and forecasted years
tick_marks = np.concatenate((actual_years, forecasted_years))
plt.xticks(tick_marks)
plt.grid(True, which='both', axis='x')

plt.title('Average Temp. Vs Year - Forecasting Set ')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# T1.2 :- (A). Finding RMSE, MAE, and MSE scores for TAVG Vs YEAR data using Linear Regression Model   
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Load the dataset
data = pd.read_csv('GTT.csv')

# Extracting the relevant columns
X = data[['YEAR']]
y = data['TAVG']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Calculate metrics
rmse = np.sqrt(mean_squared_error(y, y_pred))
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)

# Print the metrics with 8 decimal places
print(f'Root Mean Squared Error (RMSE): {rmse:.8f}')
print(f'Mean Absolute Error (MAE): {mae:.8f}')
print(f'Mean Squared Error (MSE): {mse:.8f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
import numpy as np

# Load the dataset
data = pd.read_csv('GTT.csv')

# Extracting the relevant columns
X = data[['YEAR']]
y = data['TAVG']

# Create a Lasso regression model
model = Lasso(alpha=0.1)  # Adjust alpha as needed
model.fit(X, y)

# Generate years for plotting (actual and forecasted)
years_all = np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3)
forecast_years = np.arange(data['YEAR'].max() + 1, data['YEAR'].max() + 11, 3)  # Forecast for 10 additional years with 3-year intervals

# Predictions for all years
y_pred_all = model.predict(years_all.reshape(-1, 1))
forecast_tavg = model.predict(forecast_years.reshape(-1, 1))

# Plotting time series for actual and predicted data
plt.figure(figsize=(12, 6))

# Actual data (time series plot)
plt.plot(data['YEAR'], data['TAVG'], label='Actual', color='blue')
plt.plot(years_all, y_pred_all, label='Predicted', color='red')

plt.title(' Average Temp Vs Year - Prediction Set')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.xticks(np.arange(data['YEAR'].min(), data['YEAR'].max() + 1, 3))
plt.legend()
plt.grid(True)
plt.show()

# Plotting time series for actual and forecasted data
plt.figure(figsize=(12, 6))

# Actual data (time series plot)
plt.plot(data['YEAR'], data['TAVG'], label='Actual', color='blue')
plt.plot(forecast_years, forecast_tavg, label='Forecasted', color='green')

plt.title(' Average Temp Vs Year - Forecasting Set ')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.xticks(np.arange(data['YEAR'].min(), data['YEAR'].max() + 11, 3))
plt.legend()
plt.grid(True)
plt.show()

# Plotting all three on the same graph (time series plot)
plt.figure(figsize=(12, 6))

# Actual data
plt.plot(data['YEAR'], data['TAVG'], label='Actual', color='blue')

# Prediction data (red circles)
plt.scatter(years_all, y_pred_all, label='Predicted', color='red', marker='o')

# Forecasted data (green circles)
plt.scatter(forecast_years, forecast_tavg, label='Forecasted', color='green', marker='o')

plt.title(' Average Temp Vs Year - Actual, Predicted, and Forecasted ')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.xticks(np.arange(data['YEAR'].min(), data['YEAR'].max() + 11, 3))
plt.legend()
plt.grid(True)
plt.show()
```
